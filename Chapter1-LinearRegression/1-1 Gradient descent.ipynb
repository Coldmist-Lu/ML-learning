## Gradient descent

### an easy example

求 $x^2 + 2x + 5$ 的最小值

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def func(x_ls):
    ret = []
    for x in x_ls:
        ret.append(pow(x,2) + 2*x + 5)   
    return ret
```

```python
x = np.linspace(-6, 4, 100)
y = func(x)
plt.plot(x, y)
```

```python
x_iter = 3
alpha = 0.8
iterationNum = 10
x_ls = [3]
for i in range(iterationNum):
    x_iter = x_iter - alpha*(2*x_iter + 2)
    x_ls.append(x_iter)
```

```python
print(x_iter)
plt.plot(x, y)
plt.scatter(x_ls, func(x_ls))
```

学习率调小一点

```python
x_iter = 3
alpha = 0.1 # 学习率降至0.1
iterationNum = 10
x_ls = [3]
for i in range(iterationNum):
    x_iter = x_iter - alpha*(2*x_iter + 2)
    x_ls.append(x_iter)
```

```python
print(x_iter)
plt.plot(x, y)
plt.scatter(x_ls, func(x_ls))
```
